```python
q1_area = df['area'].quantile(0.25)
q3_area = df['area'].quantile(0.75)
iqr_area = q3_area - q1_area
print(f"Area quantile 25%: {q1_area}")
print(f"Area quantile 75%: {q3_area}")
print(f"IQR for Area: {iqr_area}")
print("------------------------------------")

q1_bedrooms = df['bedrooms'].quantile(0.25)
q3_bedrooms = df['bedrooms'].quantile(0.75)
iqr_bedrooms = q3_bedrooms - q1_bedrooms
print(f"Bedrooms quantile 25%: {q1_bedrooms}")
print(f"Bedrooms quantile 75%: {q3_bedrooms}")
print(f"IQR for Bedrooms: {iqr_bedrooms}")
print("------------------------------------")

q1_bathrooms = df['bathrooms'].quantile(0.25)
q3_bathrooms = df['bathrooms'].quantile(0.75)
iqr_bathrooms = q3_bathrooms - q1_bathrooms
print(f"Bathrooms quantile 25%: {q1_bathrooms}")
print(f"Bathrooms quantile 75%: {q3_bathrooms}")
print(f"IQR for Bathrooms: {iqr_bathrooms}")
print("------------------------------------")

q1_parking = df['parking'].quantile(0.25)
q3_parking = df['parking'].quantile(0.75)
iqr_parking = q3_parking - q1_parking
print(f"Parking quantile 25%: {q1_parking}")
print(f"Parking quantile 75%: {q3_parking}")
print(f"IQR for Parking: {iqr_parking}")
```

```python
lower_bound_area = q1_area - (1.5 * iqr_area)
upper_bound_area = q3_area + (1.5 * iqr_area)
print(f"Lower bound for Area: {lower_bound_area}")
print(f"Upper bound for Area: {upper_bound_area}")
print("------------------------------------")

lower_bound_bedrooms = q1_bedrooms - (1.5 * iqr_bedrooms)
upper_bound_bedrooms = q3_bedrooms + (1.5 * iqr_bedrooms)
print(f"Lower bound for Bedrooms: {lower_bound_bedrooms}")
print(f"Upper bound for Bedrooms: {upper_bound_bedrooms}")
print("------------------------------------")

lower_bound_bathrooms = q1_bathrooms - (1.5 * iqr_bathrooms)
upper_bound_bathrooms = q3_bathrooms + (1.5 * iqr_bathrooms)
print(f"Lower bound for Bathrooms: {lower_bound_bathrooms}")
print(f"Upper bound for Bathrooms: {upper_bound_bathrooms}")
print("------------------------------------")

lower_bound_parking = q1_parking - (1.5 * iqr_parking)
upper_bound_parking = q3_parking + (1.5 * iqr_parking)
print(f"Lower bound for Parking: {lower_bound_parking}")
print(f"Upper bound for Parking: {upper_bound_parking}")
print("------------------------------------")
```

```python
from sklearn.linear_model import LinearRegression

X = df_encoded.drop(columns=["price"])
y = df_encoded["price"]

model = LinearRegression()
model.fit(X, y)

feature_importance = pd.DataFrame({"Feature": X.columns, "Importance": model.coef_})
feature_importance = feature_importance.sort_values(by="Importance", ascending=False)
print(feature_importance)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_squared_error
from sklearn.ensemble import RandomForestRegressor

# Seperate Train-Test Set (70% Train, 30% Test)
df_train, df_test = train_test_split(df_encoded, test_size=0.3, random_state=17)

# Test Set for OLS and Random Forest
y_test = df_test["price"]  # Use the test set for both models

# ========== OLS Regression ==========

# Train OLS Regression
model = smf.ols(formula='price ~ area + bathrooms + stories + parking + mainroad_yes +  guestroom_yes  + basement_yes  + hotwaterheating_yes  + airconditioning_yes  +  prefarea_yes + furnishingstatus_unfurnished', data=df_train).fit()

# summary
print(model.summary())

# Test Set predictions for OLS
y_pred_ols = model.predict(df_test)

# Calculate R² and RMSE for OLS
r2_ols = r2_score(y_test, y_pred_ols)
rmse_ols = mean_squared_error(y_test, y_pred_ols)

print(f"OLS Regression - R² on Test Set: {r2_ols:.4f}")
print(f"OLS Regression - RMSE on Test Set: {rmse_ols:.2f}")

```

```python
# ========== Random Forest ==========

# Seperate Feature and Target for Random Forest
X_train = df_train.drop(columns=["price"])
y_train = df_train["price"]
X_test = df_test.drop(columns=["price"])

# Train Random Forest
rf_model = RandomForestRegressor(n_estimators=100, random_state=17)
rf_model.fit(X_train, y_train)

# Test Set predictions for Random Forest
y_pred_rf = rf_model.predict(X_test)

# Calculate R² and RMSE for Random Forest
r2_rf = r2_score(y_test, y_pred_rf)
rmse_rf = mean_squared_error(y_test, y_pred_rf)

print(f"Random Forest - R² on Test Set: {r2_rf:.4f}")
print(f"Random Forest - RMSE on Test Set: {rmse_rf:.2f}")
```

```python
import xgboost as xgb

# ========== XGBoost Regression ==========

# Train XGBoost Model
xgb_model = xgb.XGBRegressor(n_estimators=100, learning_rate=0.1, max_depth=6, random_state=17)
xgb_model.fit(X_train, y_train)

# Test Set predictions for XGBoost
y_pred_xgb = xgb_model.predict(X_test)

# Calculate R² and RMSE for XGBoost
r2_xgb = r2_score(y_test, y_pred_xgb)
rmse_xgb = mean_squared_error(y_test, y_pred_xgb)

print(f"XGBoost - R² on Test Set: {r2_xgb:.4f}")
print(f"XGBoost - RMSE on Test Set: {rmse_xgb:.2f}")
```

```python
# ========== Compare All Models ==========
print("\n===== Model Comparison =====")
print(f"OLS Regression: R² = {r2_ols:.4f}, RMSE = {rmse_ols:.2f}")
print(f"Random Forest: R² = {r2_rf:.4f}, RMSE = {rmse_rf:.2f}")
print(f"XGBoost: R² = {r2_xgb:.4f}, RMSE = {rmse_xgb:.2f}")
```

```python
import seaborn as sns
import scipy.stats as stats

# Calculate Residuals
residuals_ols = y_test - y_pred_ols
residuals_rf = y_test - y_pred_rf
residuals_xgb = y_test - y_pred_xgb

# Set the graph size
fig, axes = plt.subplots(3, 2, figsize=(12, 12))
fig.suptitle("Residual Analysis for OLS, Random Forest, and XGBoost", fontsize=16)

# ========== Histogram ==========
sns.histplot(residuals_ols, bins=30, kde=True, ax=axes[0, 0])
axes[0, 0].set_title("OLS Residuals Histogram")

sns.histplot(residuals_rf, bins=30, kde=True, ax=axes[1, 0])
axes[1, 0].set_title("Random Forest Residuals Histogram")

sns.histplot(residuals_xgb, bins=30, kde=True, ax=axes[2, 0])
axes[2, 0].set_title("XGBoost Residuals Histogram")

# ========== Q-Q Plot ==========
stats.probplot(residuals_ols, dist="norm", plot=axes[0, 1])
axes[0, 1].set_title("OLS Q-Q Plot")

stats.probplot(residuals_rf, dist="norm", plot=axes[1, 1])
axes[1, 1].set_title("Random Forest Q-Q Plot")

stats.probplot(residuals_xgb, dist="norm", plot=axes[2, 1])
axes[2, 1].set_title("XGBoost Q-Q Plot")

plt.tight_layout(rect=[0, 0, 1, 1])  # Adjust graph spacing
plt.show()
```
